## Task 1 — Explore the Menu

```python
# printing menu grouped by category
# using .items() to get both name and details in the loop
categories = ["Starters", "Mains", "Desserts"]

for cat in categories:
    print("=====", cat, "=====")
    for name, details in menu.items():
        if details["category"] == cat:
            if details["available"] == True:
                status = "[Available]"
            else:
                status = "[Unavailable]"
            print(name, "  ₹" + str(details["price"]), " ", status)
    print()
```

```python
# using .keys() to count total items
total_items = len(menu.keys())
print("Total items:", total_items)

# using .values() to count available items
available_count = 0
for details in menu.values():
    if details["available"] == True:
        available_count = available_count + 1
print("Available items:", available_count)

# using .items() to find most expensive item
expensive_name = ""
expensive_price = 0
for name, details in menu.items():
    if details["price"] > expensive_price:
        expensive_price = details["price"]
        expensive_name = name
print("Most expensive:", expensive_name, "₹" + str(expensive_price))

# using .items() to find items under 150
print("Items under ₹150:")
for name, details in menu.items():
    if details["price"] < 150:
        print(" ", name, "₹" + str(details["price"]))
```

## Task 2 — Cart Operations

```python
cart = []

def add_item(item_name, qty):
    # check if item exists using .keys()
    if item_name not in menu.keys():
        print(item_name, "does not exist in menu")
        return
    if menu[item_name]["available"] == False:
        print(item_name, "is not available")
        return
    # if already in cart just increase qty
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = entry["quantity"] + qty
            print("Updated", item_name, "quantity to", entry["quantity"])
            return
    # not in cart so add new entry
    cart.append({"item": item_name, "quantity": qty, "price": menu[item_name]["price"]})
    print("Added", item_name, "x" + str(qty))

def remove_item(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print("Removed", item_name)
            return
    print(item_name, "not found in cart")

def update_qty(item_name, qty):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = qty
            print("Updated", item_name, "quantity to", qty)
            return
    print(item_name, "not found in cart")
```

```python
# simulating the order sequence
add_item("Paneer Tikka", 2)
print("Cart:", cart)
print()

add_item("Gulab Jamun", 1)
print("Cart:", cart)
print()

# should update to 3 not add duplicate
add_item("Paneer Tikka", 1)
print("Cart:", cart)
print()

# doesnt exist
add_item("Mystery Burger", 1)
print("Cart:", cart)
print()

# exists but unavailable
add_item("Chicken Wings", 1)
print("Cart:", cart)
print()

remove_item("Gulab Jamun")
print("Cart:", cart)
print()
```

```python
# final order summary
print("========== Order Summary ==========")

subtotal = 0
for entry in cart:
    item_total = entry["price"] * entry["quantity"]
    subtotal = subtotal + item_total
    print(entry["item"], "x" + str(entry["quantity"]), "  ₹" + str(item_total))

gst = round(subtotal * 0.05, 2)
total = round(subtotal + gst, 2)

print("------------------------------------")
print("Subtotal:       ₹" + str(subtotal))
print("GST (5%):       ₹" + str(gst))
print("Total Payable:  ₹" + str(total))
print("====================================")
```

## Task 3 — Inventory Tracker with Deep Copy

```python
import copy

# deep copy saves the original state before any changes
inventory_backup = copy.deepcopy(inventory)

# proving deep copy works - change inventory, backup should stay same
inventory["Paneer Tikka"]["stock"] = 999
print("inventory stock after change:", inventory["Paneer Tikka"]["stock"])
print("backup stock (should still be 10):", inventory_backup["Paneer Tikka"]["stock"])
print("deep copy works - backup is unaffected")
print()

# restoring inventory back to original before continuing
inventory["Paneer Tikka"]["stock"] = 10
print("Restored stock:", inventory["Paneer Tikka"]["stock"])
```

```python
# deducting cart quantities from inventory
for entry in cart:
    item = entry["item"]
    qty = entry["quantity"]
    current_stock = inventory[item]["stock"]

    if qty > current_stock:
        print("Warning: not enough stock for", item, "- deducting only", current_stock)
        inventory[item]["stock"] = 0
    else:
        inventory[item]["stock"] = current_stock - qty

print()

# reorder alerts using .items()
for name, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print("⚠ Reorder Alert:", name, "— Only", details["stock"], "unit(s) left (reorder level:", str(details["reorder_level"]) + ")")

print()
print("--- inventory now ---")
for name, details in inventory.items():
    print(name, ":", details["stock"])

print()
print("--- backup unchanged ---")
for name, details in inventory_backup.items():
    print(name, ":", details["stock"])
```

## Task 4 — Daily Sales Log Analysis

```python
# revenue per day using .items()
print("Revenue per day:")
best_day = ""
best_revenue = 0

for date, orders in sales_log.items():
    daily_total = 0
    for order in orders:
        daily_total = daily_total + order["total"]
    print(date, ": ₹" + str(daily_total))
    if daily_total > best_revenue:
        best_revenue = daily_total
        best_day = date

print()
print("Best selling day:", best_day, "₹" + str(best_revenue))
```

```python
# most ordered item - count how many orders each item appears in
item_count = {}

for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] = item_count[item] + 1

most_ordered = ""
most_count = 0

for name, count in item_count.items():
    if count > most_count:
        most_count = count
        most_ordered = name

print("Most ordered item:", most_ordered, "(", most_count, "orders )")
```
