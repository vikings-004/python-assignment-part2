## Task 1 — Explore the Menu

```python
# printing menu grouped by category
# I made a list of categories manually so they print in order
categories = ["Starters", "Mains", "Desserts"]

for cat in categories:
    print("=====", cat, "=====")
    for item in menu:
        if menu[item]["category"] == cat:
            if menu[item]["available"] == True:
                status = "[Available]"
            else:
                status = "[Unavailable]"
            print(item, "  ₹" + str(menu[item]["price"]), " ", status)
    print()
```

```python
# total items and available items
total_items = len(menu)

available_count = 0
for item in menu:
    if menu[item]["available"] == True:
        available_count = available_count + 1

print("Total items:", total_items)
print("Available items:", available_count)

# finding most expensive item
expensive_item = ""
expensive_price = 0

for item in menu:
    if menu[item]["price"] > expensive_price:
        expensive_price = menu[item]["price"]
        expensive_item = item

print("Most expensive:", expensive_item, "₹" + str(expensive_price))

# items under 150
print("Items under ₹150:")
for item in menu:
    if menu[item]["price"] < 150:
        print(" ", item, "₹" + str(menu[item]["price"]))
```

## Task 2 — Cart Operations

```python
cart = []

# function to add item to cart
def add_item(item_name, qty):
    if item_name not in menu:
        print(item_name, "does not exist in menu")
        return
    if menu[item_name]["available"] == False:
        print(item_name, "is not available")
        return
    # check if item already in cart, if yes just increase qty
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = entry["quantity"] + qty
            print("Updated", item_name, "quantity to", entry["quantity"])
            return
    # item not in cart so add new entry
    cart.append({"item": item_name, "quantity": qty, "price": menu[item_name]["price"]})
    print("Added", item_name, "x" + str(qty))

# function to remove item from cart
def remove_item(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print("Removed", item_name)
            return
    print(item_name, "not found in cart")

# function to update quantity
def update_qty(item_name, qty):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = qty
            print("Updated", item_name, "quantity to", qty)
            return
    print(item_name, "not found in cart")
```

```python
# simulating the order sequence
add_item("Paneer Tikka", 2)
print("Cart:", cart)
print()

add_item("Gulab Jamun", 1)
print("Cart:", cart)
print()

add_item("Paneer Tikka", 1)
print("Cart:", cart)
print()

add_item("Mystery Burger", 1)
print("Cart:", cart)
print()

add_item("Chicken Wings", 1)
print("Cart:", cart)
print()

remove_item("Gulab Jamun")
print("Cart:", cart)
print()
```

```python
# printing order summary
print("========== Order Summary ==========")

subtotal = 0
for entry in cart:
    item_total = entry["price"] * entry["quantity"]
    subtotal = subtotal + item_total
    print(entry["item"], " x" + str(entry["quantity"]), " ₹" + str(item_total))

gst = round(subtotal * 0.05, 2)
total = round(subtotal + gst, 2)

print("------------------------------------")
print("Subtotal:  ₹" + str(subtotal))
print("GST (5%):  ₹" + str(gst))
print("Total Payable: ₹" + str(total))
print("====================================")
```

## Task 3 — Inventory Tracker with Deep Copy

```python
import copy

# deep copy so changes to inventory dont affect the backup
inventory_backup = copy.deepcopy(inventory)

# proving deep copy works - change inventory and backup should stay same
inventory["Paneer Tikka"]["stock"] = 999
print("inventory Paneer Tikka stock:", inventory["Paneer Tikka"]["stock"])
print("backup Paneer Tikka stock:", inventory_backup["Paneer Tikka"]["stock"])
print("backup is unaffected - deep copy works!")

# restoring inventory back to original
inventory["Paneer Tikka"]["stock"] = 10
print("Restored inventory Paneer Tikka stock:", inventory["Paneer Tikka"]["stock"])
```

```python
# deducting cart items from inventory
for entry in cart:
    item = entry["item"]
    qty = entry["quantity"]
    current_stock = inventory[item]["stock"]

    if qty > current_stock:
        print("Warning: not enough stock for", item, "- deducting only", current_stock)
        inventory[item]["stock"] = 0
    else:
        inventory[item]["stock"] = current_stock - qty

print()

# reorder alerts
for item in inventory:
    stock = inventory[item]["stock"]
    reorder = inventory[item]["reorder_level"]
    if stock <= reorder:
        print("⚠ Reorder Alert:", item, "— Only", stock, "unit(s) left (reorder level:", str(reorder) + ")")

print()
print("--- inventory now ---")
for item in inventory:
    print(item, ":", inventory[item]["stock"])

print()
print("--- backup unchanged ---")
for item in inventory_backup:
    print(item, ":", inventory_backup[item]["stock"])
```

## Task 4 — Daily Sales Log Analysis

```python
# total revenue per day
print("Revenue per day:")
for date in sales_log:
    daily_total = 0
    for order in sales_log[date]:
        daily_total = daily_total + order["total"]
    print(date, ": ₹" + str(daily_total))

print()

# best selling day
best_day = ""
best_revenue = 0

for date in sales_log:
    daily_total = 0
    for order in sales_log[date]:
        daily_total = daily_total + order["total"]
    if daily_total > best_revenue:
        best_revenue = daily_total
        best_day = date

print("Best selling day:", best_day, "₹" + str(best_revenue))
```

```python
# most ordered item - count how many orders each item appears in
item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] = item_count[item] + 1

most_ordered = ""
most_count = 0

for item in item_count:
    if item_count[item] > most_count:
        most_count = item_count[item]
        most_ordered = item

print("Most ordered item:", most_ordered, "(", most_count, "orders )")
```
